```python
# Matplotlib → for plotting and showing images
%matplotlib inline
import matplotlib.pyplot as plt  

# OpenCV (cv2) → for image processing (loading, resizing, converting colors)
import cv2  
```

```python
import matplotlib.pyplot as plt
import cv2  

img = cv2.imread('cat.jpg') 

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.axis('off')
plt.title('Cat')
plt.show()
```

```python
small = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)

plt.imshow(small)
```

```python
small = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)

plt.imshow(small)
plt.axis('off')
```

```python
small = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)

plt.imshow(small)
plt.axis('off')
plt.title('Cat')
```

```python
small = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)

plt.imshow(small)
plt.axis('off')
plt.title('Cat')
plt.show()
```

```python
# Matplotlib: create a row of 3 plots
fig, axes = plt.subplots(1, 3, figsize=(12,4))

# Original
axes[0].imshow(img_rgb)
axes[0].set_title("Cat")
axes[0].axis("off")

# Grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
axes[1].imshow(gray, cmap="gray")
axes[1].set_title("Grayscale")
axes[1].axis("off")

# Resized
axes[2].imshow(small)
axes[2].set_title("Resized (100x100)")
axes[2].axis("off")

plt.show()
```
