```python
import cv2
import matplotlib.pyplot as plt

# Load image
img = cv2.imread("cat.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 1. Grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 2. Blur
blur = cv2.GaussianBlur(img_rgb, (15,15), 0)

# 3. Edge Detection
edges = cv2.Canny(img, 100, 200)

# --- Show results ---
plt.figure(figsize=(12,6))

plt.subplot(2,2,1)
plt.imshow(img_rgb)
plt.title("Original")
plt.axis("off")

plt.subplot(2,2,2)
plt.imshow(gray, cmap="gray")
plt.title("Grayscale")
plt.axis("off")

plt.subplot(2,2,3)
plt.imshow(blur)
plt.title("Blur")
plt.axis("off")

plt.subplot(2,2,4)
plt.imshow(edges, cmap="gray")
plt.title("Edges")
plt.axis("off")

plt.show()
```
